```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('climate_india_districts_output/climate_india_districts_monthly_data.csv')

plt.figure()
sns.boxplot(x='tp', data=df)
plt.xlabel('tp')
plt.title('Boxplot of tp')
plt.show()

plt.figure()
sns.boxplot(x='e', data=df)
plt.xlabel('e')
plt.title('Boxplot of e')
plt.show()

plt.figure()
sns.boxplot(x='ro', data=df)
plt.xlabel('ro')
plt.title('Boxplot of ro')
plt.show()

plt.figure()
sns.boxplot(x='t2m', data=df)
plt.xlabel('t2m')
plt.title('Boxplot of t2m')
plt.show()

plt.figure()
sns.boxplot(x='stl1', data=df)
plt.xlabel('stl1')
plt.title('Boxplot of stl1')
plt.show()

plt.figure()
sns.boxplot(x='d2m', data=df)
plt.xlabel('d2m')
plt.title('Boxplot of d2m')
plt.show()

plt.figure()
sns.boxplot(x='slt', data=df)
plt.xlabel('slt')
plt.title('Boxplot of slt')
plt.show()

plt.figure()
sns.boxplot(x='cvh', data=df)
plt.xlabel('cvh')
plt.title('Boxplot of cvh')
plt.show()

plt.figure()
sns.boxplot(x='cvl', data=df)
plt.xlabel('cvl')
plt.title('Boxplot of cvl')
plt.show()

plt.figure()
sns.boxplot(x='tcc', data=df)
plt.xlabel('tcc')
plt.title('Boxplot of tcc')
plt.show()
```
