```python
import pandas as pd
```

```python
# create a list of all csv files in the dir
import glob
import os

csv_files = glob.glob(os.path.join("*.csv"))
csv_files
```

```python
# remove the first and thrid row from each csv file
for file in csv_files:
    df = pd.read_csv(file)
    df = df.drop([1])
    df.to_csv(file, index=False)
```

```python
pd.read_csv(csv_files[0])
```

```python
all_states_districts = set()

for file in csv_files:
    df = pd.read_csv(file)
    # reading all the state and district pairs and saving them in a set
    for index, row in df.iterrows():
        all_states_districts.add((row['State'], row['District']))
    
all_states_districts = list(all_states_districts)
print(len(all_states_districts))
all_states_districts
```

```python
# convert all_state_districts to a df
df = pd.DataFrame(all_states_districts, columns=['State', 'District'])
```

```python
df.to_csv("_all_states_districts.csv", index=False)
```

```python
import geopandas as gpd
import pandas as pd

url = 'https://bharatviz.saketlab.in/India_LGD_Districts_simplified.geojson'
gdf = gpd.read_file(url)

print(gdf.columns)

state_district_df = gdf[['state_name', 'district_name']].dropna()

state_district_df = state_district_df.drop_duplicates()

state_district_df
```

```python
# editted the _all_states_districts.csv file manually to match the state and district names with the actual_states_districts.csv file

ref_df = pd.read_csv('_all_states_districts.csv')
ref_df
```

```python
state = 'Andaman and Nicobar Islands'
district = 'South andamans'

match = ref_df[(ref_df['state'] == state) & (ref_df['district'] == district)]

if not match.empty:
    corrected_state = match.iloc[0]['corrected_state']
    corrected_district = match.iloc[0]['corrected_district']
    print(corrected_state, corrected_district)
else:
    print("No match found")
```

```python
for file in csv_files:
    df = pd.read_csv(file)
    for index, row in df.iterrows():
        # see the state and district of this row
        state = row['State']
        district = row['District']
        
        match = ref_df[(ref_df['state'] == state) & (ref_df['district'] == district)]

        if not match.empty:
            corrected_state = match.iloc[0]['corrected_state']
            corrected_district = match.iloc[0]['corrected_district']
        else:
            corrected_state = state
            corrected_district = district

        # assign these new values to the df
        df.at[index, 'State'] = corrected_state
        df.at[index, 'District'] = corrected_district

    df = df.sort_values(by=['State', 'District'])

    df.to_csv(file, index=False)
```

```python
# list the files from csv_files where multiple rows have the same state and district
for file in csv_files:
    df = pd.read_csv(file)
    duplicates = df[df.duplicated(subset=['State', 'District'], keep=False)]
    if not duplicates.empty:
        print(file)
        print(duplicates)
```
